```python
def hash_md5(id):
    
    result = hashlib.md5(str(id).encode("UTF-8"))
    final_result = result.hexdigest()
    
    return final_result
```

```python
hash_md5(df_wall.loc[100, 'order_detail_id'])
```

```python
hash_md5(df_wall['order_detail_id'])
# ini artinya custom function hanya bisa melakukan 1 inputan, sehingga butuh di loop / iterate / ulang2 sepanjang data di df
```

```python
# implement costum function to whole column
# df.apply(lambda x : function_name(x['column_name']), axis = 'columns')
df_wall.apply(lambda x : hash_md5(x['order_detail_id']), axis = 'columns')
```

Ketika ingin membuat custom function dan diimplementasi ke 1 kolom di dataframe

1. coba buat corat coret function (belum dibuat def)

2. coba function yang dibuat dengan inputan salah 1 value dari df (belum dibuat def)

3. buat function menjadi general dalam bentuk def function

    """
    
        def nama_fungsi(var) :
       
           ....
        
            return output_var
   
   """
4. coba def function yang dibuat dengan memasukkan value constant dan salah 1 value dari df

5. implement function menggunakan apply lambda dengn format 

    df.apply(lambda x : function_name(x['column_name']), axis = 'columns')
