### Data Clensing

#### Remove duplication

```python
df_clean = df.copy()
```

```python
df_clean = df_clean.drop_duplicates(keep = 'last')
```

```python
df_clean[df_clean.duplicated()]
```

#### remove null row

```python
df_clean = df_clean.dropna()
```

```python
df_clean[df_clean.isna().any(axis = 1)]
```

### Replace unconsistance data

```python
# df['namacolumn'].replace({'yang_mau_direplace1' : 'hasil_replace1'})
df_clean['Gender'] = df_clean['Gender'].replace({'Male':'M', 'Female':'F'})
```

```python
list(df_clean['Gender'].unique())
```
